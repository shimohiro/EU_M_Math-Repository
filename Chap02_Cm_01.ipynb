```python
#総合問題2-1-1

import numpy as np

x = np.random.uniform(0.0, 1.0, 10000)
y = np.random.uniform(0.0, 1.0, 10000)
```

```python
#総合問題2-1-2

import math

inside_x = []
inside_y = []

outside_x = []
outside_y = []

count_inside = 0
for count in range(0,10000):
    d = math.hypot(x[count],y[count])
    if d < 1:
       count_inside += 1
       inside_x.append(x[count])
       inside_y.append(y[count])

    else:
       outside_x.append(x[count])
       outside_y.append(y[count])

print("円の内部に入った数：",count_inside)

#図
import matplotlib.pyplot as plt

plt.figure(figsize=(10,10))

circle_x = np.arange(0,1,0.001)
circle_y = np.sqrt(1- circle_x * circle_x)

plt.plot(circle_x, circle_y)

plt.scatter(inside_x,inside_y,color="cyan")
plt.scatter(outside_x,outside_y,color="chartreuse")

plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
```

```python
#総合問題2-1-3

print("円周率の近似値：",4.0 * count_inside / 10000)
```
