```python
import pandas_datareader as pdr

start_date = '2001/1/2'
end_date = '2016/12/30'

fx_jpusdata = pdr.DataReader('DEXJPUS', 'fred', start_date, end_date)
```

```python
#練習問題6-16
fx_jpusdata.resample('Y').mean().head()
```
