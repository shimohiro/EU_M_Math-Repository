```python
#練習問題5-17
import numpy as np
import scipy as sp
from scipy.optimize import fsolve

import matplotlib.pyplot as plt
%matplotlib inline

%precision 3

def f(x):
    y = 5 * x - 10
    return y

x = np.linspace(-4,4)
plt.plot(x,f(x))
plt.plot(x,np.zeros(len(x)))
plt.grid(True)
```

```python
#解を求める
x = fsolve(f,2)
print(x)
```

```python
def f2(x):
    y = x ** 3 - 2 * x ** 2 - 11 * x + 12
    return y

x = np.linspace(-5,5)
plt.plot(x,f2(x))
plt.plot(x,np.zeros(len(x)))
plt.grid(True)
```

```python
#解1を求める
x = fsolve(f2,-3)
print(x)
```

```python
#解2を求める
x = fsolve(f2,1)
print(x)
```

```python
#解3を求める
x = fsolve(f2,4)
print(x)
```
